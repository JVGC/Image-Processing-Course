```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
filename = str(input()).rstrip()
```

```python
input_img = imageio.imread("images/"+filename)
```

```python
plt.imshow(input_img, cmap="gray")
```

```python
method = int(input())
```

```python
save = int(input())
```

```python
def inversion(z):
    return 255 - z
```

```python
if method == 1:
    output_img = inversion(input_img)
rse = np.sqrt(np.sum(np.power(output_img.astype(np.int32) - input_img.astype(np.int32), 2)))
print('{:.4f}'.format(rse))
```

```python
def contrast_modulation(z, c, d):
    return (z - np.min(z))*((d - c)/(np.max(z) - np.min(z))) + c
```

```python
if method == 2:
    c = int(input())
    d = int(input())
    output_img = contrast_modulation(input_img, c, d)
output_img
```
